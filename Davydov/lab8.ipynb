```python
import numpy as np
import matplotlib.pyplot as plt
from hawkeslib import UnivariateExpHawkesProcess
```

## Симуляция процесса

```python
mu = 0.2 # интенсивность процесса
alpha = 0.15 # степень влияния предыдущих событий
beta = 0.1 # скорость затухания процесса
```

```python
uv = UnivariateExpHawkesProcess()
uv.set_params(mu, alpha, beta)
smp = uv.sample(500)

lda_ar = [mu + np.sum(alpha * beta * np.exp(-beta * (x - smp[smp < x]))) \
          for x in np.arange(0, 500, .1)]

plt.figure(figsize=(15,2))
plt.ylabel("λ*(t)")
plt.xlabel("t")
plt.plot(smp, np.ones_like(smp) * .1, 'k.')
plt.plot(np.arange(0, 500, .1), lda_ar, 'b-')
```
